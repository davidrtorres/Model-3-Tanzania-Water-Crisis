```python
"""
new_cases        new_cases_smoothed
0.0              NaN
"""

def change_value(df_data,new_cases_s, new):
    bool_series = pd.isnull(df_data[new_cases_s])
    if (bool_series.any()) & (df_data[new]==0.0): #
        df_data[new_cases_s] = 0.0
    return df_data 


new_df = change_value(df,'new_cases_smoothed','new_cases')   
```

```python
def change_valuef(df_data,new_cases_s, new ):
    bool_series = pd.isnull(df[new_cases_s])
    if (bool_series.any()) & (df_data[new]==0.0): #
        df_data[new_cases_s] = 0.0
    return df_data 
    
    
df_new = df.apply(change_value(df,'new_cases_smoothed','new_cases'), axis=1)
```
